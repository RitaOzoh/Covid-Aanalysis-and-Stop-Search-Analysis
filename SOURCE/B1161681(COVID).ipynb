```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline

def df_preprocess():
    
    global df_main
    
    print('****Required Libraries loaded****')
    df_main = pd.read_csv("specimenDate_ageDemographic-unstacked.csv", parse_dates = ["date"])
    df_main["Children"] = df_main["newCasesBySpecimenDate-0_4"] + df_main["newCasesBySpecimenDate-5_9"]
    df_main["Teenagers"] = df_main["newCasesBySpecimenDate-10_14"] + df_main["newCasesBySpecimenDate-15_19"] 
    df_main["Youths"] = df_main["newCasesBySpecimenDate-20_24"] + df_main["newCasesBySpecimenDate-25_29"] + df_main["newCasesBySpecimenDate-30_34"]
    df_main["Middle_Aged"] = df_main["newCasesBySpecimenDate-35_39"] + df_main["newCasesBySpecimenDate-40_44"] + df_main["newCasesBySpecimenDate-45_49"] + df_main["newCasesBySpecimenDate-50_54"] + df_main["newCasesBySpecimenDate-55_59"]
    df_main["Elderly"] = df_main["newCasesBySpecimenDate-60_64"] + df_main["newCasesBySpecimenDate-65_69"] + df_main["newCasesBySpecimenDate-70_74"] + df_main["newCasesBySpecimenDate-75_79"] + df_main["newCasesBySpecimenDate-80_84"] + df_main["newCasesBySpecimenDate-85_89"] + df_main["newCasesBySpecimenDate-90+"]
    df_main['Unassigned'] = df_main['newCasesBySpecimenDate-unassigned']
    df_main['Young_Group'] = df_main['newCasesBySpecimenDate-0_59']
    df_main['Elderly_Group'] = df_main['newCasesBySpecimenDate-60+']
    df_main["Total_Cases"] = df_main['Young_Group'] + df_main['Elderly_Group']
    cols_to_drop = ['newCasesBySpecimenDate-0_4', 'newCasesBySpecimenDate-5_9',
       'newCasesBySpecimenDate-10_14', 'newCasesBySpecimenDate-15_19',
       'newCasesBySpecimenDate-20_24', 'newCasesBySpecimenDate-25_29',
       'newCasesBySpecimenDate-30_34', 'newCasesBySpecimenDate-35_39',
       'newCasesBySpecimenDate-40_44', 'newCasesBySpecimenDate-45_49',
       'newCasesBySpecimenDate-50_54', 'newCasesBySpecimenDate-55_59',
       'newCasesBySpecimenDate-unassigned', 'newCasesBySpecimenDate-60_64',
       'newCasesBySpecimenDate-65_69', 'newCasesBySpecimenDate-70_74',
       'newCasesBySpecimenDate-75_79', 'newCasesBySpecimenDate-80_84',
       'newCasesBySpecimenDate-85_89', 'newCasesBySpecimenDate-90+',
       'newCasesBySpecimenDate-0_59', 'newCasesBySpecimenDate-60+',
       'newCasesBySpecimenDateRollingRate-0_4',
       'newCasesBySpecimenDateRollingRate-5_9',
       'newCasesBySpecimenDateRollingRate-10_14',
       'newCasesBySpecimenDateRollingRate-15_19',
       'newCasesBySpecimenDateRollingRate-20_24',
       'newCasesBySpecimenDateRollingRate-25_29',
       'newCasesBySpecimenDateRollingRate-30_34',
       'newCasesBySpecimenDateRollingRate-35_39',
       'newCasesBySpecimenDateRollingRate-40_44',
       'newCasesBySpecimenDateRollingRate-45_49',
       'newCasesBySpecimenDateRollingRate-50_54',
       'newCasesBySpecimenDateRollingRate-55_59',
       'newCasesBySpecimenDateRollingRate-60_64',
       'newCasesBySpecimenDateRollingRate-65_69',
       'newCasesBySpecimenDateRollingRate-70_74',
       'newCasesBySpecimenDateRollingRate-75_79',
       'newCasesBySpecimenDateRollingRate-80_84',
       'newCasesBySpecimenDateRollingRate-85_89',
       'newCasesBySpecimenDateRollingRate-90+',
       'newCasesBySpecimenDateRollingRate-0_59',
       'newCasesBySpecimenDateRollingRate-60+',
       'newCasesBySpecimenDateRollingSum-0_4',
       'newCasesBySpecimenDateRollingSum-5_9',
       'newCasesBySpecimenDateRollingSum-10_14',
       'newCasesBySpecimenDateRollingSum-15_19',
       'newCasesBySpecimenDateRollingSum-20_24',
       'newCasesBySpecimenDateRollingSum-25_29',
       'newCasesBySpecimenDateRollingSum-30_34',
       'newCasesBySpecimenDateRollingSum-35_39',
       'newCasesBySpecimenDateRollingSum-40_44',
       'newCasesBySpecimenDateRollingSum-45_49',
       'newCasesBySpecimenDateRollingSum-50_54',
       'newCasesBySpecimenDateRollingSum-55_59',
       'newCasesBySpecimenDateRollingSum-unassigned',
       'newCasesBySpecimenDateRollingSum-60_64',
       'newCasesBySpecimenDateRollingSum-65_69',
       'newCasesBySpecimenDateRollingSum-70_74',
       'newCasesBySpecimenDateRollingSum-75_79',
       'newCasesBySpecimenDateRollingSum-80_84',
       'newCasesBySpecimenDateRollingSum-85_89',
       'newCasesBySpecimenDateRollingSum-90+',
        'newCasesBySpecimenDateRollingSum-0_59',
       'newCasesBySpecimenDateRollingSum-60+']
    df_main = df_main.drop(cols_to_drop, axis = 1)
    df_sorted = df_main.sort_values(by='date')
    df_sorted_ind = df_sorted.set_index('date')
    return df_sorted_ind
    
def date_filter_range(df,from_date, to_date):
    
    mask = (df.index >= from_date) & (df.index <= to_date)
    df_range = df.loc[mask]
    
    return df_range

def date_filter_single(df,day):
    
    mask = (df.index == day)
    df_range = df.loc[mask]
    return df_range

def groupby_fxn(df, group=''):
    """
    This function takes the dataframe
    group = The column to groupby    
   """
    df_grouped = df.groupby(df[group], as_index = True).sum()
    
    return df_grouped

def sort_fxn(df, sort_by=''):
    
    with pd.option_context('mode.use_inf_as_null', True):
        df_sorted = df.sort_values(sort_by, ascending=False, axis=0, na_position='last').head(5)
    return df_sorted

def plot_line(x, y, x_label='', y_label='', title=''):
    import matplotlib.pyplot as plt
    import seaborn as sns
    sns.set()
    fig, ax = plt.subplots(figsize=(15,6), dpi=80)
    plt.plot(x, y, marker='o', markerfacecolor='black', markersize=8)
    plt.xlabel(x_label, size=12)
    plt.ylabel(y_label, size=12)
    plt.title(title, size=15)
    plt.xticks(rotation=90)
    for index in range(len(x)):
      ax.text(x[index], y[index], y[index], size=12)
    plt.show()

def plot_bar_norm(df, x_label='', title=''):
    import matplotlib.pyplot as plt
    import seaborn as sns
    sns.set()
    ax1 = df.plot(kind='bar', figsize=(20, 8), color="indigo", fontsize=13)
    ax1.set_alpha(0.8)
    ax1.set_title(title, fontsize=21)
    ax1.set_xlabel(x_label, fontsize=18)
    for i in ax1.patches:
        ax1.annotate(format(i.get_height()), (i.get_x() + i.get_width() / 2., i.get_height()), ha='center',
                     va='center', xytext=(0, 10), textcoords='offset points', fontsize=12)
    plt.show()
    
def plot_bar_pct_change(df, x_label='', title=''):
    import matplotlib.pyplot as plt
    import seaborn as sns
    sns.set()
    ax1 = df.plot(kind='bar', figsize=(20, 8), color="indigo", fontsize=13)
    ax1.set_alpha(0.8)
    ax1.set_title(title, fontsize=21)
    ax1.set_xlabel(x_label, fontsize=18)
    for i in ax1.patches:
        height = i.get_height()
        ax1.text(x=i.get_x() + i.get_width() / 2, y=height+.10,
        s="{}%".format(height), ha='center', fontsize=12)
    plt.show()
    
    
def date_report_display(df,freq, from_date, to_date):
    """ freq = {daily, weekly, monthly}
        date_format = yyyy-mm-dd
    """
    df_range = date_filter_range(df,from_date, to_date)
    if str.lower(freq) == 'daily':
        df_sorted_daily = df_range.resample('D').sum()
        x = df_sorted_daily.index.tolist()
        y = df_sorted_daily.Total_Cases.tolist()
        plot_line(x, y, x_label='Days', y_label='Total_Count', title='Total number of covid-19 cases per day')
        
        
    elif str.lower(freq) == 'weekly':
        df_sorted_weekly = df_range.resample('W').sum()
        x = df_sorted_weekly.index.isocalendar().week.tolist()
        y = df_sorted_weekly.Total_Cases.tolist()
        plot_line(x, y, x_label='Week', y_label='Total_Count', title='Total number of covid-19 cases per Week')
        
        
    elif str.lower(freq) == 'monthly':
        df_sorted_monthly = df_range.resample('M').sum()
        x = df_sorted_monthly.index.month_name().tolist()
        y = df_sorted_monthly.Total_Cases.tolist()
        plot_line(x, y, x_label='Month', y_label='Total_Count', title='Total number of covid-19 cases per Month')
        
        
def feature_report_display_single_day(df, day, group_col, plot_col):
    print(day)
    df_range = date_filter_single(df,day)
    df_range = df_range[[group_col, plot_col]]
    df_day = sort_fxn(groupby_fxn(df_range, group=group_col), sort_by=plot_col)
    plot_bar_norm(df_day, x_label=group_col, title=f'{group_col} With Highest number of Covid-19 Cases')
    
    
def feature_date_range_change(df, from_date, to_date, group_col, change_col, plot_col='Percentage_Change'):
    import numpy as np
    df_range = date_filter_range(df,from_date, to_date)
    df_range[plot_col] = np.round(df_range[change_col].pct_change()* 100, 2)
    df_range = df_range[[group_col, plot_col]]
    df_day = sort_fxn(groupby_fxn(df_range, group=group_col), sort_by=plot_col)
    plot_bar_pct_change(df_day, x_label=group_col, title=f'{group_col} With the Largest positive percentage changes in Covid-19 Cases')
```
